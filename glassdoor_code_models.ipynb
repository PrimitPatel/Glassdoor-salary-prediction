```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('glassdoor_data_preprocessed.csv')
df
```

```python
df.columns
```

```python
cor=df.corr()
cor
```

```python
cor_target = abs(cor["avg_salary"])
cor_target.sort_values()
```

```python
df_model = df[['avg_salary','Size','Type of ownership','Industry','Sector','Revenue','no_of_comp','Hourly_salary','Employer_provided_salary','state','same_state','Python','R','Spark','SQL','AWS','Excel','Tableau','PowerBi','job_category','seniority','description_length']]
df_dum = pd.get_dummies(df_model)
df_dum 
```

```python
# get the values for axis ----- categories on x and avg salary on y
X = df_dum.drop('avg_salary', axis =1)
Y=df_dum.avg_salary.values
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=  train_test_split(X,Y,test_size=0.2, random_state= 42)
```

```python
# multiple linear regression
from sklearn import linear_model
ml_regr = linear_model.LinearRegression()
ml_regr.fit(X_train, Y_train)

#multiple linear regression evaluation
cross_val_score_ml =np.mean(cross_val_score(ml_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3))
print('Mean error for Ml',cross_val_score_ml)
```

```python
# Lasso regressor
from sklearn.linear_model import Lasso

alpha = []
error = []
for i in range(1,100):
    alpha.append(i/100)
    Lasso_regr = Lasso(alpha=(i/100))
    error.append(np.mean(cross_val_score(Lasso_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
  
df_err = pd.DataFrame(zip(alpha,error), columns = ['alpha','error'])
print(df_err[df_err.error == max(df_err.error)])
```

```python
Lasso_regr = Lasso(alpha=0.06)
Lasso_regr.fit(X_train,Y_train)
print('mean error for lasso: ',np.mean(cross_val_score(Lasso_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
```

```python
from sklearn.linear_model import Ridge
alpha = []
error = []
for i in range(1,100):
    alpha.append(i/100)
    Ridge_regr = Ridge(alpha=(i/100))
    error.append(np.mean(cross_val_score(Ridge_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
    
df_err = pd.DataFrame(zip(alpha,error), columns = ['alpha','error'])
print(df_err[df_err.error == max(df_err.error)])
```

```python
Ridge_regr = Ridge(alpha=.99)
Ridge_regr.fit(X_train,Y_train)
print('mean error for Ridge: ',np.mean(cross_val_score(Ridge_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
```

```python
# Support vector regression
from sklearn.svm import SVR
svr_regr=SVR(kernel='poly')
svr_regr.fit(X_train,Y_train)

print('mean error for SVR: ',np.mean(cross_val_score(svr_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
```

```python
# decision tree regresson
from sklearn.tree import DecisionTreeRegressor
ds_regr = DecisionTreeRegressor(random_state = 0)
ds_regr.fit(X_train, Y_train)
print('mean error for Decision tree: ',np.mean(cross_val_score(ds_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
```

```python
# random forest regresson
from sklearn.ensemble import RandomForestRegressor
rf_regr = RandomForestRegressor()
rf_regr.fit(X_train,Y_train)
print('mean error for Random Forest Regressor: ',np.mean(cross_val_score(rf_regr,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
```

```python
rf1 = RandomForestRegressor()
parameters = {'n_estimators':range(10,300,10), 'criterion':('mse','mae'), 'max_features':('auto','sqrt','log2')}


#Tuning the random forest regression with RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
clf = RandomizedSearchCV(rf1, parameters, n_iter=170, cv=3, random_state=1)
clf.fit(X_train, Y_train)
```

```python
clf.best_estimator_
```

```python
print('mean error for Random Forest Regressor after tuning: ',np.mean(cross_val_score(clf.best_estimator_,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
```

```python
from sklearn.model_selection import GridSearchCV
ds1 = DecisionTreeRegressor()
parameters_ds = { 'criterion':('mse','mae','friedman_mse'), 'max_features':('auto','sqrt','log2'),'splitter':('best', 'random'),}
#Tuning the random forest regression with RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
clf1 = GridSearchCV(ds1, parameters_ds,cv=3)
clf1.fit(X_train, Y_train)

```

```python
clf1.best_estimator_
```

```python
print('mean error for Ds after tuning: ',np.mean(cross_val_score(clf1.best_estimator_,X_train,Y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
```

```python
ml_pred = ml_regr.predict(X_test)
lasso_pred= Lasso_regr.predict(X_test)
ridge_pred= Ridge_regr.predict(X_test)
svr_pred= svr_regr.predict(X_test)
ds_pred= ds_regr.predict(X_test)
rf_pred= rf_regr.predict(X_test)
clf_pred = clf.best_estimator_.predict(X_test)
clf_pred1 = clf1.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print('MAE for ml >',mean_absolute_error(Y_test,ml_pred))
print('MAE for lasso >',mean_absolute_error(Y_test,lasso_pred))
print('MAE for ridge >',mean_absolute_error(Y_test,ridge_pred))
print('MAE for svr >',mean_absolute_error(Y_test,svr_pred))
print('MAE for ds >',mean_absolute_error(Y_test,ds_pred))
print('MAE for rf >',mean_absolute_error(Y_test,rf_pred))
print('MAE for rf after tuning >',mean_absolute_error(Y_test,clf_pred))
print('MAE for ds after tuning >',mean_absolute_error(Y_test,clf_pred1))
```
